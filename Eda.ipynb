# CountPlot

```python
sns.countplot(x ='sex',data = data,color ='green',hue = 'sex',palette='seismic')
plt.title("CountPlot Of Sex")
```

* The chart shows that there are slightly more males than females in the dataset

* The chart displays a slightly higher count for males compared to females

* In this chart it is said that the male customer are getting the insurance

* The chart shows that the number of males and females is almost equal,The nearly equal heights of the bars suggest a balanced dataset in terms of gender

# Histogram

```python
sns.histplot(x='age',bins=20,data=data,kde=True,color='salmon')
plt.title("HistPlot Of Age")
```

* The age group around 20 has the highest count,indicating that the age of 20 peoples are applied for the insurance

* The count gradually decreases after age 60,This Indicate The age of 60 are not applied the insurance

# LinePlot

```python
sns.lineplot(x = 'age',y = 'children',data = data,color='skyblue')
plt.title("LinePlot Of Age Vs Children")
```

* The plot shows a general trend of fewer children as age increases, with significant individual variability

# ScatterPlot

```python
sns.scatterplot(x='age',y='bmi',data=data,hue = 'bmi',palette = 'coolwarm')
plt.title("ScatterPlot Of Age Vs Bmi")
```

* The plot covers a wide age range, likely from approximately 20 to 60 years old

* The scatterplot shows a wide range of BMIs across all age groups, with no clear frequency patterns based on age

# BarPlot

```python
sns.barplot(x='region', y='bmi', data=data,hue='region')
plt.title("BarPlot Of Region Vs Bmi")
```

* The bar plot shows that BMI varies across regions, with the southeast region having the highest average BMI and the northwest region having the lowest



# BoxPlot

```python
sns.boxplot(x='bmi', y='sex', data=data,hue='sex')
plt.title("BoxPlot Of Bmi Vs Sex")
```

 * The boxplot shows that BMI varies by sex, with females having a slightly lower median BMI than males, but with more outliers indicating a wider range of BMI values

# HeatMap

```python
correlation_matrix =df1[['age','bmi','children']].corr()
print('correlation matrix:')
print(correlation_matrix)


sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidth=0.5)

plt.title("HeatMap Of Age,Bmi,Children")
```

* The heatmap shows a weak positive correlation between age and BMI, and a very weak correlation between age and children, while BMI and children have a negligible correlation

# PieChart

```python
labels = ['southwest','southeast','northeast','northwest']
sizes = [27.9,33.8,33.0,22.7]  

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=140)
plt.legend(title = "Region:")
```

* Southeast has the largest share at 28.8%, indicating it is the most significant region in terms of the variable being measured

* Northwest follows closely with 28.1%, making it almost as significant as the southeast

* Southwest has a substantial share as well at 23.8%, suggesting it also plays a considerable role

* Northeast has the smallest share at 19.3%, indicating it is the least significant of the four regions in this context
